```python
# From the unique values, we can see that the first column is the sentiment and the last column is the tweet text.
# So, I will rename the columns accordingly.

# Two columns are important in this dataset [0] and [6]

def set_columns_names(dataframe):
    # Get number of columns
    n_cols = len(dataframe.columns)
    
    new_column_names = []
    
    for i in range(n_cols):
        # print(i)
        new_column_names.append(f"col_{i}")
        
    return new_column_names

```

```python
cols = set_columns_names(df)
```

```python
# Rename Columns
df.columns = cols
```

```python
# Get important columns only col_0 and col_5

df = df[['col_0','col_5']]
```

```python
df.columns = ['lable','twits']
```

```python
cleaner = cln(df)
df['clean_twits'] = df['twits'].apply(cleaner.clean_text)
```
